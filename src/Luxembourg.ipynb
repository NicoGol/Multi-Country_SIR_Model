```python
from plot_functions import *
Horizon = 350
def compare_infect_curve(countries,country,Ires,names,ymax=None):
    plots = []    
    j = countries.index(country)
    pop = df_countries['population'][country]
    for i in range(len(Ires)):
        plots.append(go.Scatter(y=np.array(Ires[i][j])/pop, name="Infected "+names[i]))
    fig = go.Figure(data=plots, layout=go.Layout(barmode='group'))
    fig.update_layout(template="plotly_white", margin=dict(l=0, r=0, t=30, b=0),width=1000,height=500,title='Comparaison of infected curves in different scenarios in '+country)
    if ymax is not None:
        fig.update_layout(yaxis_range=[0.0,ymax])
    fig.show()
    return fig
```

```python
def reduce_beta_country(country,init_beta=0.2,new_beta=0.15,t_change_beta=30,gamma=0.1,Horizon=350,origin='united kingdom',t_change_mig=30,ymax=None):

    countries = create_countries(eu_countries, origin=origin, beta=init_beta, gamma=gamma, Horizon=Horizon)
    countries[eu_countries.index(country)].beta[t_change_beta:] = [new_beta for i in range(t_change_beta,Horizon)]
    mig = create_migration_matrix(Horizon) *1000
    Ires1 = simulate(countries, mig, Horizon=Horizon)

    countries = create_countries(eu_countries, origin=origin, beta=init_beta, gamma=gamma, Horizon=Horizon)
    countries[eu_countries.index(country)].beta[t_change_beta:] = [new_beta for i in range(t_change_beta,Horizon)]
    mig = create_migration_matrix(Horizon) *1000
    mig = modify_mig('all','united kingdom',eu_countries,0.1,mig,t=t_change_mig)
    Ires2 = simulate(countries,mig, Horizon=Horizon) 

    countries = create_countries(eu_countries, origin=origin, beta=init_beta, gamma=gamma, Horizon=Horizon)
    countries[eu_countries.index(country)].beta[t_change_beta:] = [new_beta for i in range(t_change_beta,Horizon)]
    mig = create_migration_matrix(Horizon) *1000
    mig = modify_mig(country,'all',eu_countries,0.1,mig,t=t_change_mig)
    Ires3 = simulate(countries,mig, Horizon=Horizon) 

    return compare_infect_curve(eu_countries,country,[Ires1,Ires2,Ires3],['baseline', 'when isolating UK at time '+str(t_change_mig), 'when isolating '+country+' at time '+str(t_change_mig)],ymax=ymax)
```

```python
fig = reduce_beta_country('luxembourg',Horizon=400,new_beta=0.2)
```
